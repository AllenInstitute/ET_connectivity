```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# load files
pre_df = pd.read_csv('L5-ET-pre-motif-percentage.csv')
post_df = pd.read_csv('L5-ET-post-motif-percentage.csv')

# data on X-axis
pre_mean = pre_df.iloc[:,1:].mean()
pre_sem = pre_df.iloc[:,1:].sem()
post_mean = post_df.iloc[:,1:].mean()
post_sem = post_df.iloc[:,1:].sem()
```

```python
# plot
fig = plt.figure(figsize=(10,6))

#plt.figure(figsize=(10,6))

N = 20
ind = np.arange(N)

width = 0.35       
plt.ylim([0,60])

plt.grid(axis = 'y',linestyle = '--', linewidth = 0.5, zorder=1)

plt.bar(ind, pre_mean, width, label='Synapses from presynaptic Motif groups onto postsynaptic ET cells', color='lightsteelblue', zorder=2)
plt.bar(ind + width, post_mean, width, label='Synapses from presynaptic ET cells onto postsynaptic Motif groups',color='darkblue', zorder=2)

yerr1=pre_sem
yerr2=post_sem

plt.vlines( ind, pre_mean, pre_mean+yerr1, color='black')
plt.vlines( ind+ width, post_mean, post_mean+yerr2, color='black')

plt.xlabel('Motif group #', fontsize=15)
plt.ylabel('Percentage of synapse', fontsize=15)
#plt.title('Data from 39 ETs',fontsize=18)

plt.xticks(ind + width / 2, (np.arange(1,21)), fontsize=12)
plt.yticks(fontsize=12)

plt.legend(loc='best',framealpha=0,fontsize=14,bbox_to_anchor=(0.65, 0.65))


for pos in ['right', 'top']:
    plt.gca().spines[pos].set_visible(False)

fig.savefig('Extended_Fig2.eps', bbox_inches='tight')
    
plt.show()
```
